```python
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='whitegrid')
```

```python
df_house = pd.read_csv("housing.csv")
df_house.head()
```

```python
df_value_income = df_house[['median_house_value', 'median_income']]
df_value_income
g = sns.relplot(data=df_value_income, x="median_house_value", y="median_income")
```

```python
model = linear_model.LinearRegression()
```

### Train the model

```python
# fit
model.fit(X=df_house[["median_income"]], y=df_house[["median_house_value"]])
```

### Evaluate the model

```python
# print coefficients
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")
print(f"Prediction: {model.predict(X=[[12],[1],[100]])}")
```

```python
12*model.coef_ + model.intercept_
```

```python
sns.scatterplot(x="median_income", y="median_house_value", data=df_house, color="blue")
plt.plot(df_house[["median_income"]], model.predict(df_house[["median_income"]]), color="black", linewidth=3)
plt.show()
```
