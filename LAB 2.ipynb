## Graph 1: Matplotlib

This graph is created using _Matplotlib_ and shows a sine wave with values ranging from 0 to 10.

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plotting using Matplotlib
plt.plot(x, y)
plt.title('Matplotlib Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## Graph 2: Seaborn

The _Seaborn_ scatter plot demonstrates a linear relationship with random noise between X and Y.

```python
import seaborn as sns
import numpy as np

# Example data
x = np.random.randn(100)
y = 3 * x + np.random.randn(100)

# Creating a DataFrame for Seaborn
import pandas as pd
data = pd.DataFrame({'X': x, 'Y': y})

# Plotting using Seaborn
sns.scatterplot(data=data, x='X', y='Y')
plt.title('Seaborn Random Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## Graph 3: Plotly

The _Plotly_ line chart visualizes an exponential curve with increasing values along the X-axis.

```python
import plotly.express as px
import numpy as np

# Example data
x = np.linspace(0, 17, 100)
y = np.exp(x)

# Plotting using Plotly
fig = px.line(x=x, y=y, labels={'x': 'X-axis', 'y': 'Y-axis'}, title='Plotly Exponential Curve')
fig.show()
```
